# Imports

```python
from utils import *
import pandas as pd
import numpy as np
import datetime
import os

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import *
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

os.chdir('C:/Users/howar/Desktop/Coursework/ME4-Individual-Project-Volatility-Forecasting')
```

# LSTM

## Data processing

```python
# Intraday returns
r = pd.read_csv('data/intraday_returns.csv', index_col=0, parse_dates=True)

# Compute covariance matrix
N = len(r.columns)
covariance = compute_covariance(r).values.reshape((-1, N, N))

# Square root -> Cholesky decomposition -> vech
cov_vectors = decompose_covariance(covariance)

# Generate sequences for LSTM
window = 60
X, y = generate_sequences(cov_vectors, window=window, additional_features=False)
X.shape, y.shape
```

```python
# Test over 2021
r_test = pd.read_csv('data/daily_returns_test.csv', index_col=0, parse_dates=True)
test_window = len(r_test)

# Don't shuffle test data to keep sequence order
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_window, shuffle=False)

# Shuffle train and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, shuffle=True, random_state=42)

X_train.shape, X_val.shape, X_test.shape
```

## Define and run model

```python
inputs = Input(X_train.shape[1:])
x = LSTM(32, activation='tanh', return_sequences=True, recurrent_regularizer=l2(5e-3))(inputs)
x = LSTM(64, activation='tanh', return_sequences=False, recurrent_regularizer=l2(5e-3))(x)
x = Dense(128, activation='relu')(x)
x = Dense(256, activation='relu')(x)
outputs = Dense(y_train.shape[1], activation='linear')(x)

model = Model(inputs, outputs)
model.compile(optimizer=Adam(3e-4), loss='mse')

es = EarlyStopping('val_loss', patience=100, restore_best_weights=True)

history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    epochs=10000,
                    batch_size=64,
                    verbose=2,
                    callbacks=[es])
```

## LSTM results

### Training, validation and testing errors

```python
lstm_train_pred = model.predict(X_train)
lstm_train_mse = np.mean(np.square(y_train - lstm_train_pred))

lstm_val_pred = model.predict(X_val)
lstm_val_mse = np.mean(np.square(y_val - lstm_val_pred))

lstm_test_pred = model.predict(X_test)
lstm_test_mse = np.mean(np.square(y_test - lstm_test_pred))

print('Training MSE:', lstm_train_mse)
print('Validation MSE:', lstm_val_mse)
print('Testing MSE:', lstm_test_mse)
```

### Portfolio optimization

```python
# Undo vectorization and decomposition
covariance = []
for vec in lstm_test_pred:
    covariance.append(reconstruct_covariance(vec))
    
covariance = np.array(covariance)

# Create portfolio and run optimization
portfolio = Portfolio(r_test, covariance)
results = portfolio.optimize(return_percent=True)

# Print performance metrics and plot history
portfolio.metrics(starting_value=100) 
```

### Save

```python
model.save('outputs/LSTM.h5')
np.save('outputs/LSTM/covariance_forecasts.csv', covariance)
portfolio.weights.to_csv('outputs/LSTM/portfolio_weights.csv', index=True)
```

# LSTM Sentiment

## Data processing

```python
r = pd.read_csv('data/intraday_returns.csv', index_col=0, parse_dates=True)
sentiment = pd.read_csv('data/sentiment_features.csv', index_col=0, parse_dates=True)

# Compute covariance matrix
N = len(r.columns)
cov = compute_covariance(r)

# Square root -> Cholesky decomposition -> vech
cov_vectors = decompose_covariance(cov.values.reshape((-1, N, N)))

# Generate sequences for LSTM
window = 60
Xcov, Xsent, y = generate_sequences(cov_vectors, features=sentiment, additional_features=True, window=window)
Xcov.shape, Xsent.shape, y.shape
```

```python
Xcov_train, Xcov_test, y_train, y_test = train_test_split(Xcov, y, test_size=test_window, shuffle=False)
Xsent_train, Xsent_test = train_test_split(Xsent, test_size=test_window, shuffle=False)

Xcov_train, Xcov_val, y_train, y_val = train_test_split(Xcov_train, y_train, test_size=0.2, shuffle=True, random_state=42)
Xsent_train, Xsent_val = train_test_split(Xsent_train, test_size=0.2, shuffle=True, random_state=42)

print('No. training samples:', len(y_train))
print('No. of validation samples:', len(y_val))
print('No. testing samples:', len(y_test))
```

## Define and run model

```python
# Covariance model
inputs_cov = Input(Xcov_train.shape[1:])
xcov = LSTM(32, activation='tanh', return_sequences=True, recurrent_regularizer=l2(5e-3))(inputs_cov)
xcov = LSTM(64, activation='tanh', return_sequences=False, recurrent_regularizer=l2(5e-3))(xcov)
outputs_cov = Dense(128, activation='relu')(xcov)
model_cov = Model(inputs_cov, outputs_cov)

# Yield curve model
inputs_sent = Input((None, Xsent_train.shape[2]))
xsent = LSTM(32, activation='tanh', return_sequences=False, recurrent_regularizer=l2(5e-3))(inputs_sent)
outputs_sent = Dense(64, activation='relu')(xsent)
model_sent = Model(inputs_sent, outputs_sent)

x = tf.concat([model_cov.outputs, model_sent.outputs], axis=-1)
x = Lambda(lambda x: tf.squeeze(x, [0]))(x)
x = Dropout(0.2)(x)
outputs = Dense(y_train.shape[-1], activation='linear')(x)

model = Model([inputs_cov, inputs_sent], outputs)
model.compile(optimizer=Adam(3e-4), loss='mse')

es = EarlyStopping('val_loss', patience=100, restore_best_weights=True)

history = model.fit([Xcov_train, Xsent_train], y_train,
                    validation_data=([Xcov_val, Xsent_val], y_val),
                    epochs=10000, batch_size=64,
                    verbose=2, callbacks=[es])
```

## LSTM Sentiment results

### Training, validation and testing errors

```python
lstm_sent_train_pred = model.predict([Xcov_train, Xsent_train])
lstm_sent_train_mse = np.mean(np.square(y_train - lstm_sent_train_pred))

lstm_sent_val_pred = model.predict([Xcov_val, Xsent_val])
lstm_sent_val_mse = np.mean(np.square(y_val - lstm_sent_val_pred))

lstm_sent_test_pred = model.predict([Xcov_test, Xsent_test])
lstm_sent_test_mse = np.mean(np.square(y_test - lstm_sent_test_pred))

print('Training MSE:', lstm_sent_train_mse)
print('Validation MSE:', lstm_sent_val_mse)
print('Testing MSE:', lstm_sent_test_mse)
```

### Portfolio optimization

```python
# Undo vectorization and decomposition
covariance = []
for vec in lstm_sent_test_pred:
    covariance.append(reconstruct_covariance(vec))
    
covariance = np.array(covariance)

# Create portfolio and run optimization
portfolio = Portfolio(r_test, covariance)
results = portfolio.optimize(return_percent=True)

# Print performance metrics and plot history
portfolio.metrics(starting_value=100) 
```

### Save

```python
model.save('outputs/LSTM Sentiment.h5')
np.save('outputs/LSTM Sentiment/covariance_forecasts.csv', covariance)
portfolio.weights.to_csv('outputs/LSTM Sentiment/portfolio_weights.csv', index=True)
```

# LSTM Sentiment-Yield Curve

## Data processing

```python
r = pd.read_csv('data/intraday_returns.csv', index_col=0, parse_dates=True)
sentiment = pd.read_csv('data/sentiment_features.csv', index_col=0, parse_dates=True)
yc = pd.read_csv('data/yield_curves.csv', index_col=0, parse_dates=True)

# Compute covariance
cov = compute_covariance(r)

# Match dates
yc = yc[yc.index.isin(cov.index.get_level_values(0))]
cov = cov[cov.index.get_level_values(0).isin(yc.index.date)]
sentiment = sentiment[sentiment.index.isin(cov.index.get_level_values(0))]

# Compute covariance vectors
N = len(r.columns)
cov_vectors = decompose_covariance(cov.values.reshape((-1, N, N)))

# Create sequences
window = 60
Xcov, Xsent, Xyield, y = [], [], [], []

for i in range(len(cov_vectors) - window):
    Xcov.append(cov_vectors[i:i + window])
    Xsent.append(sentiment.iloc[i:i + window].values)
    Xyield.append(yc.iloc[i:i + window].values)
    y.append(cov_vectors[i + window])

Xcov = np.array(Xcov)
Xsent = np.array(Xsent)
Xyield = np.array(Xyield)
y = np.array(y)

Xcov.shape, Xsent.shape, Xyield.shape, y.shape
```

```python
# Train-test-split
Xcov_train, Xcov_test, y_train, y_test = train_test_split(Xcov, y, test_size=test_window, shuffle=False)
Xsent_train, Xsent_test = train_test_split(Xsent, test_size=test_window, shuffle=False)
Xyield_train, Xyield_test = train_test_split(Xyield, test_size=test_window, shuffle=False)

# Train-val-split
Xcov_train, Xcov_val, y_train, y_val = train_test_split(Xcov_train, y_train, test_size=0.2, shuffle=True, random_state=42)
Xsent_train, Xsent_val = train_test_split(Xsent_train, test_size=0.2, shuffle=True, random_state=42)
Xyield_train, Xyield_val = train_test_split(Xyield_train, test_size=0.2, shuffle=True, random_state=42)

print('No. training samples:', len(y_train))
print('No. of validation samples:', len(y_val))
print('No. testing samples:', len(y_test))
```

## Define and run model

```python
# Covariance model
inputs_cov = Input((None, Xcov_train.shape[2]))
xcov = LSTM(32, activation='tanh', return_sequences=False, recurrent_regularizer=l2(5e-3))(inputs_cov)
outputs_cov = Dense(64, activation='relu')(xcov)
model_cov = Model(inputs_cov, outputs_cov)

# Sentiment model
inputs_sent = Input((None, Xsent_train.shape[2]))
xsent = LSTM(32, activation='tanh', return_sequences=False, recurrent_regularizer=l2(5e-3))(inputs_sent)
outputs_sent = Dense(64, activation='relu', kernel_regularizer=l2(5e-3))(xsent)
model_sent = Model(inputs_sent, outputs_sent)

# yc model
inputs_yield = Input((None, Xyield_train.shape[2]))
xyield = LSTM(32, activation='tanh', return_sequences=False, recurrent_regularizer=l2(5e-3))(inputs_yield)
outputs_yield = Dense(64, activation='relu', kernel_regularizer=l2(5e-3))(xyield)
model_yield = Model(inputs_yield, outputs_yield)

x = tf.concat([model_cov.outputs, model_sent.outputs, model_yield.outputs], axis=-1)
x = Lambda(lambda x: tf.squeeze(x, [0]))(x)
x = Dropout(0.2)(x)
outputs = Dense(y_train.shape[-1], activation='linear')(x)

model = Model([inputs_cov, inputs_sent, inputs_yield], outputs)
model.compile(optimizer=Adam(3e-4), loss='mse')

es = EarlyStopping('val_loss', patience=100, restore_best_weights=True)

history = model.fit([Xcov_train, Xsent_train, Xyield_train], y_train,
                    validation_data=([Xcov_val, Xsent_val, Xyield_val], y_val),
                    epochs=10000, batch_size=64,
                    verbose=2, callbacks=[es])
```

## Results

### Training, validation and testing errors

```python
lstm_sent_yield_train_pred = model.predict([Xcov_train, Xsent_train, Xyield_train])
lstm_sent_yield_train_mse = np.mean(np.square(y_train - lstm_sent_yield_train_pred))

lstm_sent_yield_val_pred = model.predict([Xcov_val, Xsent_val, Xyield_val])
lstm_sent_yield_val_mse = np.mean(np.square(y_val - lstm_sent_yield_val_pred))

lstm_sent_yield_test_pred = model.predict([Xcov_test, Xsent_test, Xyield_test])
lstm_sent_yield_test_mse = np.mean(np.square(y_test - lstm_sent_yield_test_pred))

print('Training MSE:', lstm_sent_yield_train_mse)
print('Validation MSE:', lstm_sent_yield_val_mse)
print('Testing MSE:', lstm_sent_yield_test_mse)
```

### Portfolio optimization

```python
# Undo vectorization and decomposition
covariance = []
for vec in lstm_sent_yield_test_pred:
    covariance.append(reconstruct_covariance(vec))
    
covariance = np.array(covariance)

# Create portfolio and run optimization
portfolio = Portfolio(r_test, covariance)
results = portfolio.optimize(return_percent=True)

# Print performance metrics and plot history
portfolio.metrics(starting_value=100) 
```

### Save

```python
model.save('models/LSTM Sentiment Yield.h5')
np.save('outputs/LSTM Sentiment Yield/covariance_forecasts.csv', covariance)
portfolio.weights.to_csv('outputs/LSTM Sentiment Yield/portfolio_weights.csv', index=True)
```
